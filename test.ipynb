```python
import pandas as pd

df = pd.read_csv("data/edges_with_scores_readable.csv")
print("Num unique nurses:", df['nurse_id'].nunique())
print("Num unique shifts:", df['shift_node'].nunique())
print("Num unique wards:", df['ward'].unique())
print(df.groupby('ward')['shift_node'].nunique())

# For each nurse and fortnight, sum possible hours
df['week'] = pd.to_datetime(df['date']).dt.isocalendar().week
df['fortnight'] = (df['week'] - df['week'].min()) // 2
coverage = df.groupby(['nurse_id', 'fortnight'])['duration_hours'].sum().reset_index()
print(coverage.sort_values('duration_hours'))
```
